```python
import matplotlib.pyplot as pit

x= [1,2,3,4,5]
y = [1,4,9,16,25]

# create a line plot

pit.plot(x,y)
pit.xlabel('X axis')
pit.ylabel('Y axis')
pit.title('Basic line plot')

```

```python
x= [1,2,3,4,5]
y = [1,4,9,16,25]

# Create customized line plot

pit.plot(x,y,color='red', linestyle='--', marker='o', linewidth=3, markersize=9)
pit.grid(True)
```

```python
# Multiple Plots

x= [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]

pit.figure(figsize=(9,5))
pit.subplot(1,2,1)
pit.plot(x,y1,color='green')
pit.title('Plot 1')

pit.subplot(1,2,2)
pit.plot(x,y2,color='green')
pit.title('Plot 2')
```

```python
category = ['A', 'B', 'C', 'D', 'E']
value = [1,22,33,51,6]

pit.bar(category, value, color='purple')
```

Histograms:- Histograms are used to represent the distribution of a dataset. They divide the data into bins and count the number of data points in each bin

```python
data = [1,2,3,3,4,4,5,6,7,8,8,9]

# create a histogram
pit.hist(data, bins=5)
```

```python
# create a scatter plot

x =[1,2,3,4,5]
y = [2,3,4,5,6]

pit.scatter(x,y, color='blue', marker='x')
```

```python
# create pipe chart

labels = ['A','B', 'C', 'D']
sizes = [30,20,40,10]
colors=['gold', 'yellowgreen','lightcoral', 'lightskyblue']

explode = (0.2,0,0,0) #move out the first slice

pit.pie(sizes,explode=explode ,labels= labels, colors=colors, autopct="%1.1f%%", shadow=True)
```

```python
# Sales Data Visualization
import pandas as pd

sales_data = pd.read_csv('dummy.csv')
sales_data.head(5)
```

```python
sales_data.info()
```

```python
# plot total sales by products
total_sales_by_product = sales_data.groupby('Category')['Total Sales'].sum()
print(total_sales_by_product)
```

```python
total_sales_by_product.plot(kind='bar', color='teal')
```

```python
#plot sales trend over time

sales_trend = sales_data.groupby('Date')['Total Sales'].sum().reset_index()
pit.plot(sales_trend['Date'], sales_trend['Total Sales'])
```
